```python


from nba_api.stats.endpoints import boxscoretraditionalv3
import pandas as pd

game_id = '0022500492' 

trad_box = boxscoretraditionalv3.BoxScoreTraditionalV3(game_id=game_id)

player_trad = trad_box.get_data_frames()[0]

team_trad = trad_box.get_data_frames()[1]

player_trad.to_csv("player_traditional_stats.csv", index=False)
team_trad.to_csv("team_traditional_stats.csv", index=False)

print("✅ Traditional stats saved to CSV files:")
print("   - player_traditional_stats.csv")
print("   - team_traditional_stats.csv")
```

```python


from nba_api.stats.endpoints import boxscoreadvancedv3

adv_box = boxscoreadvancedv3.BoxScoreAdvancedV3(game_id=game_id)

player_adv = adv_box.get_data_frames()[0]

team_adv = adv_box.get_data_frames()[1]

player_adv.to_csv("player_advanced_stats.csv", index=False)
team_adv.to_csv("team_advanced_stats.csv", index=False)

print("✅ Advanced stats saved to CSV files:")
print("   - player_advanced_stats.csv")
print("   - team_advanced_stats.csv")
```
